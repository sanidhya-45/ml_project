```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

# Load the Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Preprocess the data
x_train = x_train / 255.0
x_test = x_test / 255.0
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)

# Split the data into train and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Set k-value for k-NN
k = 5

# Create a k-NN classifier
knn = KNeighborsClassifier(n_neighbors=k)

# Fit the k-NN model on the training data
knn.fit(x_train, y_train)

# Predict on the validation data
y_val_pred = knn.predict(x_val)

# Calculate validation accuracy
val_accuracy = accuracy_score(y_val, y_val_pred)

# Predict on the test data
y_test_pred = knn.predict(x_test)

# Calculate test accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print validation accuracy and test accuracy
print("Validation Accuracy: {:.2f}%".format(val_accuracy * 100))
print("Test Accuracy: {:.2f}%".format(test_accuracy * 100))

# Create a confusion matrix for validation data
cm_val = tf.math.confusion_matrix(labels=y_val, predictions=y_val_pred).numpy()

# Create a confusion matrix for test data
cm_test = tf.math.confusion_matrix(labels=y_test, predictions=y_test_pred).numpy()

# Plot the confusion matrices
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(cm_val, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Validation Data)")
plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Test Data)")
plt.show()

# Plot the accuracy graph
k_values = range(1, 21)
val_accuracies = []
test_accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_val_pred = knn.predict(x_val)
    y_test_pred = knn.predict(x_test)
    val_acc = accuracy_score(y_val, y_val_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    val_accuracies.append(val_acc)
    test_accuracies.append(test_acc)

plt.plot(k_values, val_accuracies, label='Validation Accuracy')
plt.plot(k_values, test_accuracies, label='Test Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('k-NN Validation and Test Accuracy')
plt.legend()
plt.show()
```
